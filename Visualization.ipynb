```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
#To get the built-in files
sns.get_dataset_names()
```

```python
#Loading a particular dataset from the above to a particular dataset
df = sns.load_dataset('mpg')
```

```python
df.head() #rows are observationd and columns are the features
```

```python
df.shape
```

398 rows and 9 columns

```python
df.isnull().sum()
```

Eliminating the null values

```python
df = df.dropna()
```

Univariate Analysis :

```python
#plotting histogram fro one variable
px.histogram(df , x='mpg')
#telling us the most common values i.e. frequency of the values
```

```python
#plotting box plot for visualising median , outliers and quartiles
#px.box(df,x='mpg')
px.box(df,y='mpg')
```

```python
#plotting categorical values
px.histogram(df,x='origin' , color='origin')
```

```python
#pie chart
px.pie(df, color='origin' ,names='origin')
```

```python
#plotting histogram and box plot
px.histogram(df,x='mpg' , color='origin' , marginal='box')
```

```python
#categorical + continuous variables = Box plots
px.box(df,y ='mpg', color = 'origin')
```

```python
px.histogram(df , x = 'origin' , color='cylinders' , barnorm= 'percent')
```

Most are 6 cylinder cars in USA followed by 4 cylinder cars , followed by 6 cylinder cars
Japan has majority of 4 cylinder cars , followed by 6 cylinder cars and followed by 3 cylinder cars
europe has 4 cylinder cars in majority  , followed by 6 and 5 cylinders respectively  

```python
#grouping these stacked bars for a grouped view #
px.histogram(df, x = 'origin', color = 'cylinders', barnorm = 'percent', barmode = 'group')
```

```python
#two continous varibales - scatter plot
px.scatter(df,x='horsepower' , y = 'mpg')
```

Horsepower is inversly proportional to miles per gallons(mpg)

```python
px.scatter(df, x= 'horsepower' , y='mpg', color='origin')
```

It can be inferred that , Japan has more dominance of mpg and horsepower , whereas , cars produced in usa are not doing that great

```python
#size of the dots
px.scatter(df, x= 'horsepower' , y='mpg', color='origin' , size='mpg')
```

The small dots represent less mpg whereas bigger dots represent more mpg

```python
#different coloumns
px.scatter(df,x='mpg',y = 'horsepower' , color='origin',size='mpg',facet_col='origin')
```

Shows distribution in three different countries in different columns for better understanding of the visualisation

```python
#Trend line
px.scatter(df, x= 'horsepower' , y='mpg', trendline='ols')
```

Gives a regression line

```python
#3d
px.scatter_3d(df, x = 'weight', y = 'horsepower', z = 'mpg')#, color = 'cylinders', size = 'displacement')
```

```python
px.scatter_3d(df, x = 'weight', y = 'horsepower', z = 'mpg', color = 'cylinders', size = 'displacement')
```

```python
#scatter matrix - which can accomadate many variables  -Pair plot
px.scatter_matrix(df, dimensions= ['mpg', 'horsepower', 'displacement', 'weight'], color = 'origin')
```
